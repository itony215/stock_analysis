```python
import pandas as pd
import numpy as np
import os
df=pd.read_excel("./data/1605.xlsx")
df.insert(5, "漲跌幅", 0) 
df['漲跌幅'] = round(df['收盤價'].pct_change()*100,2)
```

```python
target=[]
for i in df.index:
    if not(df.loc[i].isnull().values.any()):
        if((df.loc[i]['漲跌幅']<-7)):  #跌超過7%日期抓出來觀察
            target.append(i-1)
            target.append(i)
target
```

```python

oo = '1605'
new_df = pd.DataFrame(df, index= target)
new_df.insert(0, "stock_id",oo)
res = pd.concat([new_df,new_df],axis=0)
display(res)
```

```python
with os.scandir('./data/') as it:
    combine=pd.DataFrame([])
    for entry in it:
        if entry.name.endswith(".xlsx") and entry.is_file():
            print(entry.name, entry.path)
            df=pd.read_excel(entry.path)
            df.insert(5, "漲跌幅", 0) 
            df['漲跌幅'] = round(df['收盤價'].pct_change()*100,2)
            target=[]
            stock_id = entry.name.split(".")[0]
            for i in df.index:
                if not(df.loc[i].isnull().values.any()):
                    #if((df.loc[i]['漲跌幅']<-7)):  #跌超過 7% 日期抓出來觀察
                    if((df.loc[i]['收盤價']-df.loc[i]['開盤價'])/df.loc[i]['開盤價']<-0.05):  #當日跌幅超過 5% 前一天資料
                        target.append(i-1)
                        #target.append(i)
            new_df = pd.DataFrame(df, index= target)
            new_df.insert(0, "stock_id",stock_id)
        combine = pd.concat([combine,new_df],axis=0)
combine.to_csv("./filter.csv",encoding='utf_8_sig', index=False)
```
